#Project 2: NYC Dept. of Education Data on NYS Math Exam results

Data source: https://nycopendata.socrata.com/data?cat=education  
Data description:  
NYS Math Exam Results for NYC between Grades 3-8 from 2006-2012.  
Level 1- Well below proficient in standards for their grade, insufficient for the expectations at this grade  
Level 2- Below level in standards for their grade, partial but insufficient  
Level 3- Proficient in standards for their grade, sufficient for the expecttations at this grade   
Level 4- Excel in standards for their grade, ore than sufficient for the expectations at this grade  

_These definitions are set by the NYS Exam Board_



###What am I exploring?  
- Exam Proficiency levels across all of NYC  
- Exam Proficiency levels within boroughs  
- Exam Proficiency levels across all of NYC by gender  
- Exam Proficiency levels in NYC changes from 2006-2011

##Female vs Male G3-8 Students at AVERAGE Math proficiency (2006-2012)

```python
N = 7
BoysLevel3 = Allboys.groupby(['Year']).sum().Level3
GirlsLevel3 = Allgirls.groupby(['Year']).sum().Level3

ind = np.arange(N)
width = 0.35

fig, ax = plt.subplots()
rectsl = ax.bar(ind, BoysLevel3, width, color='b')
rects2 = ax.bar(ind+width, GirlsLevel3, width, color='y')



# add some
ax.set_ylabel('Number of Students')
ax.set_title('Number of Students at Standard Math Proficiency Level for their Grade', fontsize='15')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('2006', '2007', '2008', '2009', '2010', '2011', '2012') )

ax.legend( ('rects1'[0], 'rects2'[0]), ('Boys', 'Girls') )

ax.legend(('Boys', 'Girls'), loc='upper right')
```
