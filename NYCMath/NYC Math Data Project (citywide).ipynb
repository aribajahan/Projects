```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
NYCMath = pd.read_csv("MathTestCitywide.csv")
```

```python
NYCMath
```

```python
#New data frame with only the info I care about
FixedMathdf = NYCMath [['Grade', 'Year', 'Number Tested', 'Level1', 'Level2', 'Level3', 'Level4']]
FixedMathdf
```

```python

FixedMathdf.set_index(['Year'])
#index is applied to x axis
plt.figure(figsize=(12,5))

plt.scatter(FixedMathdf.Year, FixedMathdf.Level3, lw=10, alpha=.1)
#whats the alpha?
plt.xlabel("Year")
plt.ylabel("Level3")
plt.title("Proficieny at Level 3 over time",fontsize='15')

#set_index('Year')
```
